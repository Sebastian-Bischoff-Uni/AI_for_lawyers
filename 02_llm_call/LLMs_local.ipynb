#### Einfache Anfrage an lokales Modell über ```ollama```

Beispielcode findet sich in der offiziellen [Dokumentation](https://github.com/ollama/ollama-python) von ```Ollama-Python```

````python
from ollama import chat

response = chat(model='gemma3:1b', messages=[ # gibt ein Response-Objekt zurück
  {
    'role': 'user',
    'content': 'kann ich durch Kaufvertrag Eigentum erwerben im deutschen Recht?',
  },
])

# Ausgabe der Response-Objekts
print(response)
# direkter Zugriff auf die relevanten Felder des Response-Objekts
print(response.message.content)
# print(response['message'], ['content'])
````

````python
from ollama import chat

stream = chat(
    model='gemma3:27b',
    messages=[{'role': 'user', 'content': 'kann ich durch Kaufvertrag Eigentum erwerben im deutschen Recht?'}],
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
````
