## Python-Methoden

### Wichtigste Python-Standard-Funktionen

pop()

Entfernt und gibt das letzte Element der Liste zurück (oder ein Element an einem bestimmten Index).

```python
lst = [10, 20, 30, 40]
x = lst.pop()      # entfernt
y = lst.pop(1)     
print(lst)         # [10, 30]
```

append()  --> Wichtig!

Fügt ein Element am Ende der Liste hinzu.

```python
lst = [1, 2]
lst.append(3)
print(lst)  # [1, 2, 3]
```

count()

Zählt, wie oft ein bestimmter Wert in der Liste vorkommt.

```python
lst = [1, 2, 2, 3]
print(lst.count(3))  # 2
```

extend()

Erweitert die Liste um alle Elemente einer anderen Iterable (z. B. Liste).

```python
lst = [1, 2]
lst.extend([3, 4])
print(lst)  # [1, 2, 3, 4]
```

insert()

Fügt ein Element an einer bestimmten Position (Index) ein.

```python
lst = [1, 3]
lst.insert(1, 2)
print(lst)  # [1, 2, 3]
```

remove()

Entfernt das **erste** Vorkommen eines bestimmten Wertes aus der Liste.

```python
lst = [1, 2, 2, 3]
lst.remove(2)
print(lst)  # [1, 2, 3]
```

min() / max()

Gibt das kleinste bzw. größte Element eines Iterable zurück.

```python
lst = [5, 2, 9]
print(min(lst))  # 2
print(max(lst))  # 9
```

join()

Verbindet alle Strings einer Iterable zu einem einzigen String, wobei der String vor .join() als Trennzeichen verwendet wird.

```python
woerter = ["Ich", "lerne", "Python"]
satz = " ".join(woerter)
print(satz)  # Ich lerne Python
```

range() 

Erzeugt eine Zahlenfolge, die oft in Schleifen verwendet wird, zum Beispiel um einen Codeblock mehrmals auszuführen.

```python
for i in range(0, 5, 2):
    print(i)
```

enumerate()

Ermöglicht es, beim Durchlaufen einer Liste, gleichzeitig Index und Wert zu erhalten. Sie kreiert dabei Tupel.

```python
print(list(enumerate("abc")))
```

```python
liste = ["a", "b", "c"]

for index, wert in enumerate(liste):
    print(index, wert)
```

zip()

Verbindet mehrere Iterables (z. B. Listen) zu Paaren, sodass ihre Elemente gleichzeitig durchlaufen werden können.

```python
liste_1 = [1, 2, 3]
liste_2 = ["a", "b", "c"]
x = list(zip(liste_1, liste_2))
print(x)
```
