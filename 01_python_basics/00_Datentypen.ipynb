```python
#Integer:
print ( 7 , -77 , 0 ) # ganze Zahlen, können positiv oder negativ sein, aber keine Nachkommastellen haben

#Float:
print ( 7.0 , -77.0 , 0.0 ) # Gleitkommazahlen, können positiv oder negativ sein, müssen aber Nachkommastellen haben

#String:
print ( 'Hello World' , "Hello World" ) # Zeichenketten, können in einfachen oder doppelten Anführungszeichen geschrieben werden, um Sonderzeichen zu vermeiden

# Boolean:
print ( True , False ) # Wahrheitswerte, können nur die Werte True oder False annehmen, sie sind wichtig für die Steuerung von Programmabläufen und die Logik von Bedingungen

# None:
print ( None ) # None ist ein spezieller Datentyp, der den Wert "nichts" repräsentiert, er wird oft verwendet, um das Fehlen eines Wertes oder eines Objekts anzuzeigen
```

```python
gruß = "Hallo Tim"
print(gruß)
```

```python
# Tupel: 
tupel = ( 1 , 2.0 , 'Hello World', True ) # Tupel sind unveränderlich, können aber verschiedene Datentypen enthalten
print ( tupel , tupel[0] , tupel[1] , tupel[2] ) # Zugriff auf die Elemente eines Tupels mit Indexierung, beginnend bei 0
```

```python

# Listen:
liste = [ 1 , 2.0 , 'Hello World' ] # Listen sind veränderlich, können aber verschiedene Datentypen enthalten
print ( liste , liste[0] , liste[1] , liste[2] ) # Zugriff auf die Elemente einer Liste mit Indexierung, beginnend bei 0
```

```python
dict = { 'eins' : 1 , 'zwei' : 2.0 , 'drei' : 'Hello World' } #  sind veränderlich, können aber verschiedene Datentypen enthalten
print ( dict , dict['eins'] , dict['zwei'] , dict['drei'] ) # Zugriff auf die Elemente eines Dictionaries mit Schlüssel, der in diesem Fall ein String ist
print ( dict.keys ( ) ) # Extraktion der keys
print ( dict.values ( ) ) # Extraktion der values
print ( dict.items ( ) ) # Extraktion der key-value-Paares
```

```python
# sets:
set = { 1 , 2.0 , 'Hello World' } # Sets sind veränderlich, können aber verschiedene Datentypen enthalten, sie sind ungeordnet und enthalten keine Duplikate
print ( set ) # Zugriff auf die Elemente eines Sets ist nicht möglich, da sie ungeordnet sind
```

```python
# Escape-Sequenzen:
print ( 'Das ist eine Zeile.\nDas ist eine neue Zeile.' ) # \n ist die Escape-Sequenz für einen Zeilenumbruch
print ( 'Das ist eine Zeile.\tDas ist eine neue Zeile.' ) # \t ist die Escape-Sequenz für einen Tabulator
print ( 'Das ist eine Zeile.\\ Das ist eine neue Zeile.' ) # \\ ist die Escape-Sequenz für einen Backslash, damit wird der Backslash als normales Zeichen interpretiert und nicht als Beginn einer Escape-Sequenz
```

```python
#verschachtelte Datenstrukturen: (nested data structures)
verschachtelte_liste = [ 1 , [ 2.0 , 'Hello World' ] , { 'eins' : 1 , 'zwei' : 2.0 , 'drei' : 'Hello World' } ] # eine Liste, die eine andere Liste und ein Dictionary enthält
print( verschachtelte_liste[2]['eins'] )
```

```python
# Überprüfen der Datentypen:
print ( type ( '' ) , type ( True ) , type ( 0 ) , type ( 0. ) )
print ( type ( [] ) , type ( () ) , type ( {} ) )
```
