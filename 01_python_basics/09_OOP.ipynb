## Objektorientierte Programmierung (Object-Oriented Programming, OOP)

Objektorientierte Programmierung ist ein Programmierstil, bei dem Programme aus Objekten aufgebaut werden, die sowohl Daten (Attribute) als auch Verhalten (Methoden/Funktionen) enthalten.

Man kann sich das so vorstellen: Statt nur einzelne Befehle auszuführen, modelliert man Dinge wie „Auto“, „Person“ oder „Konto“ als eigene Einheiten im Code – ähnlich wie in der realen Welt.

Konzept: 
1. es gibt Objekte
2. diese haben Eigenschaften
3. und Fähigkeiten

vgl. dazu [Real Python](https://realpython.com/python3-object-oriented-programming/?utm_source=chatgpt.com)

```python

class Student:
	# Attribute, Eigenschaften
    def __init__(self, name, alter, noten):
            self.vorname = name[0] # class-attribute
            self.nachname = name[1]
            self.alter = alter
            self.noten = noten
        # Methoden:
    def showdata (self): #self ist stets erster Parameter!
        print (self.vorname, self.nachname, self.alter) 
		# self erlaubt den Zugriff auf etwas außerhalb der Funktion, aber innerhalb der class
    
    def durchschnitt(self):
        wert = sum(self.noten) / len(self.noten)
        return wert

    def bewertung(self):
        avg = self.durchschnitt()
        if avg <= 2:
            wert = "gut"
        else:
            wert = "nicht gut"
        return wert

# Test der neuen class:

ich = Student(['Tim', 'Leibold'], 26, [1, 2, 2]) # ein Objekt / eine Instanz einer class instanziieren
ich.showdata() 
print(ich.durchschnitt())
print(ich.bewertung())
```
