## Variablen

Variablen sind Speicherplätze im RAM, wo man Daten ablegen kann und immer das Datum liegt, das man als letztes dort abgelegt hat.


Variablen ermöglichen es, Daten unter einem Namen zu speichern und später wiederzuverwenden, statt immer mit festen Zahlen oder Werten zu arbeiten. Dadurch wird Code deutlich verständlicher, weil die Namen (z. B. preis_apfel) erklären, wofür ein Wert steht. Außerdem macht es Programme flexibler, da man Änderungen nur an einer Stelle vornehmen muss und nicht im gesamten Code.

```python
preis_apfel = 0.5
preis_brot = 2.0
print(preis_apfel == preis_brot)
anzahl_aepfel = 4
anzahl_brote = 2

gesamtpreis = preis_apfel * anzahl_aepfel + preis_brot * anzahl_brote

print("Gesamtpreis:", gesamtpreis, "Euro")
```

```python
# Wertzuweisungskette
a = b = c = d = 7
print ( a , b , c , d ,'\n\n', a)
```
