## Operatoren

#### Arithmetische Operatoren

```python
a = 10
b = 3

print(a + b)   
print(a - b)  
print(a * b)  
print(a / b) # 3.3333333333333335 (Division)
print(a // b)  # 3 (Ganzzahlige Division)
print(a % 2)   # 1  (Modulo/Rest)
print(a ** b)  # 1000 (Potenz)
```

#### Vergleichsoperatoren

```python
print(5 == 5)   # True  (gleich)
print(5 != 3)   # True  (ungleich)
print(5 > 3)    # True  (größer)
print(5 < 3)    # False (kleiner)
print(5 >= 5)   # True  (größer gleich)
print(5 <= 4)   # False (kleiner gleich)
```
