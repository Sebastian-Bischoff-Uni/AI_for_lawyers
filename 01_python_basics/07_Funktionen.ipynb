## Funktionen 

Mithilfe von Funktionen kann man bestimmte Aufgaben an einer Stelle definieren und dann beliebig oft aufrufen.
+ Funktionen helfen dabei, Programme übersichtlicher, kürzer zu schreiben und besser zu strukturieren
+ Funktionen sind insbesondere wiederverwendbar
+ Debugging wird einfacher

Statt viele einzelne Schritte direkt in das Hauptprogramm zu schreiben, kann man einzelne Arbeitsschritte des Workflows auf verschiedene Funktionen auslagern.
Beispiel:
Funktion A: Berechnet den Altersdurchschnitt
Funktion B: Schreibt eine Begrüßungs-E-Mail 
Funktion C: Prüft eine Eingabe


##### Aufbau:
+	**Definition**: "def" leitet die Funktionsdefinition ein
+	**Parameter**: () enthält mögliche Eingaben
+	**Aufruf**: der eingerückte Block gehört zur Funktion

##### Einfachstes Beispiel:

Defintion

```python
def begruessung():
    print("Hallo!")
```

Parameter

```python
def begruessung(name):
    print(f"Hallo, {name}!")
```

Aufruf

```python
begruessung("Tim")
```

### Funktionen mit und ohne Rückgabewert

Werden Funktionen über den return-Befehl mit der Fähiigkeit ausgestattet, einen bestimmten Ausgabewert zurükzugeben, do kann automatisiert mit deren Rückgabewert weitergearbbeitet werden

1. Funktion ohne return

```python
def sage_hallo(name):
    print("Hallo", name)

sage_hallo("Tim")
```

2. Funktion mit return

```python
def berechne_durchschnitt(liste):
    summe = 0
    for zahl in liste:
        summe += zahl
        durchschnitt = summe / len(liste)
    return durchschnitt

zahlen = [2, 4, 6, 8]
print(berechne_durchschnitt(zahlen))
```

#### Kombintation mit Schleifen 

```python
def quadrat(x):
    return x ** 2

zahlen = [1, 2, 3, 4, 5]

for zahl in zahlen:
    print(quadrat(zahl))
```
