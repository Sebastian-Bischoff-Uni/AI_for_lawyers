## If Else Elif

Mit if, elif und else kannst du dein Programm Entscheidungen treffen lassen.
Der Code wird also nur dann ausgeführt, wenn bestimmte Bedingungen erfüllt sind.

```python
zahl = 7

if zahl > 10:   # erste Voraussetzung
    print("größer als 10")
elif zahl > 5:  # weiterer Voraussetzung
    print("größer als 5")
else:           # Wenn keine der Bedingungen zutrifft
    print("5 oder kleiner")
```

```python
personen = {
    "Sophia" : 4,
    "Anna": 17,
    "Ben": 22,
    "Clara": 35,
    "David": 70
}

for name, alter in personen.items():
    
    if alter < 7:
        kategorie = "geschäftsunfähig"
    elif alter < 18:
        kategorie = "beschränkt geschäftsfähig"
    else:
        kategorie = "unbeschränkt geschäftsfähig"

    
    print(name, "ist", kategorie)
```
