## Schleifen (Loops)

Schleifen führen zu einem Abweichen von der zeilenweise Verarbeitung.

Loops ermöglichen es, Code automatisch mehrfach auszuführen, ohne ihn jedes Mal neu schreiben zu müssen.
Dadurch wird dein Code kürzer, effizienter und leichter anpassbar – besonders wenn du viele ähnliche Aufgaben hast (z. B. Listen durchgehen).

##### While-Loops

Vergleichsoperatoren:  <  <=    >  >=    ==     !=   (P2: <>)

Schlüsselwort while, danach Bedingung, gefolgt von Doppelpunkt.
Nach dem Doppelpunkt sind die zugehörigen Folgezeilen systematisch eingerückt.

Code wird so lange ausgeführt, wie eine definierte **Bedingung** den Wert True hat (Bool!)

```python
i = 0

while i < 5:
    print(i)
    i += 1
```

```python
x = 0
while x < 5:
    print(f'x ist aktuell bei {x}')
    x = x + 1 # alternativ x += 1
else:
    print("x ist nicht kleiner als 5")
```

##### For-Loops 

Eine for-Schleife durchläuft **alle Elemente einer Sammlung** (z. B. Liste oder Dictionary).

```python
for i in range(5):
    print(i)
```

```python
namen = ["Anna", "Ben", "Clara"]

for name in namen:
    print(name)
```

```python
schueler = {
    "Anna": [2, 1, 3],
    "Ben": [3, 4, 2],
    "Clara": [1, 1, 2],
    "David": [4, 5, 3]
}

for name, noten in schueler.items(): # mit der .item() Funktion werden die Dictionaries in Tupel mit Key und Value umgewandelt
    summe = 0
    
    for note in noten:
        summe += note # gleichbedeutend zu summme = summe + note
    
    durchschnitt = summe / len(noten) # mit len(noten) wird die Anzahl der Einzelnoten in der Notenliste ermittelt
    
    if durchschnitt <= 2:
        bewertung = "sehr gut"
    elif durchschnitt <= 3:
        bewertung = "gut"
    elif durchschnitt <= 4:
        bewertung = "ausreichend"
    else:
        bewertung = "mangelhaft"
    
    print(name, "hat den Durchschnitt", durchschnitt, "->", bewertung)
    # print(f"{name} hat den Nurchschnitt {durchschnitt} -> {bewertung}")
```
